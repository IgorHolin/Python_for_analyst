```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Лекции/Churn_Modelling.csv')
df.head()
```

### Стандартный синтаксис создания графиков

```python
plt.figure(figsize=(8,6))
plt.plot([0,1,2,3],[10,9,6,1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('example')
plt.grid()
```

```python
plt.figure(figsize=(8,8))
plt.hist(df['Age'], bins=15)
plt.title('Age of clients')
plt.xlabel('Years')
plt.ylabel('Quantity');
```

```python
plt.figure(figsize=(8,8))
sns.histplot(df['Age'], bins=15)
plt.title('Age of clients')
plt.xlabel('Years')
plt.ylabel('Quantity');
```
