```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./laptops.csv', sep=';')
df.head()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split(' ')[0])
df
```

```python
df['Memory_amount'] = df['Memory'].apply(lambda x: x.split(' ')[0])
df
```

```python
def trans(x):
    if x.endswith('GB'):
        res = x.replace('GB', '')
    elif x.endswith('TB'):
        res = float(x.replace('TB',''))*1024
    return int(res)

df['Memory_amount'] = df['Memory_amount'].apply(trans)
df
```

```python
df['Memory_amount'].value_counts()
```

```python
df['Memory_type'] = df['Memory'].apply(lambda x: ' '.join(x.split(' ')[1:]))
df
```

```python
df['Memory_type'].value_counts()
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df
```

```python
df['SSD'] = 0
df
```

```python
df.loc[df['Memory_type'] == 'SSD', 'SSD'] = 1
df
```

```python
df['Weight'] = df['Weight'].apply(lambda x: x.replace('kg', '')).astype('float')
df
```
